```python
import pandas as pd
import requests
import json

url = 'http://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList'

서비스키 = 'qaWpm78A25TbAEGjCatvHW6BdFus1G5fye9RUl9DnSHJmCLKzr9A4yNE4TQApY8/kPEoCEai2OcTN+BMg1JJLw=='

# df_weather_concatenated = pd.DataFrame()

start_date = '20231001'
end_date = '20231101'

date = start_date
while date <= end_date:
 
    params ={'serviceKey' : f'{서비스키}', 'pageNo' : '1', 'numOfRows' : '100', 'dataType' : 'json', 'dataCd' : 'ASOS', 'dateCd' : 'HR', 'startDt' : f'{date}', 'startHh' : '00', 'endDt' : f'{date}', 'endHh' : '23', 'stnIds' : '152' }

    response = requests.get(url, params=params)
    data_dict = json.loads(response.content)
   
    if 'item' in data_dict['response']['body']['items']:
        df_weather = pd.DataFrame(data_dict['response']['body']['items']['item'])
        df_weather = df_weather[['tm', 'ta', 'rn', 'ws', 'hm', 'ss', 'icsr']]
        df_weather_concatenated = pd.concat([df_weather_concatenated, df_weather], ignore_index=True)
    
    date = (pd.to_datetime(date) + pd.DateOffset(days=1)).strftime('%Y%m%d')
```

```python
df_weather_concatenated = df_weather_concatenated.rename(columns={
    'tm': 'useDate',
    'ta': 'temp',
    'rn': 'rain',
    'ws': 'wind',
    'hm': 'hum',
    'ss': 'ss',
    'icsr': 'sr',
})
```

```python
df_weather_concatenated = df_weather_concatenated.replace('',0)
```

```python
df_weather_concatenated.to_csv(r'data/weather.csv')
```
